# ----- Rank of a Matrix -----

## *From Scratch (Row Echelon Form Method)*

```python
import numpy as np
```

```python
def rank_from_scratch(A):
    A = A.astype(float)
    rows, cols = A.shape
    rank = 0

    for r in range(rows):
        if A[r][r] != 0:
            # Make pivot 1
            A[r] = A[r] / A[r][r]
            for i in range(rows):
                if i != r:
                    A[i] = A[i] - A[i][r] * A[r]
        else:
            # Swap with a lower row that has a non-zero pivot
            for i in range(r+1, rows):
                if A[i][r] != 0:
                    A[[r, i]] = A[[i, r]]
                    break
    # Count non-zero rows
    for row in A:
        if not np.allclose(row, 0):
            rank += 1
    return rank
```

```python
# ----- Example -----
A = np.array([[2, 4, 1],
              [6, 12, 3],
              [1, 2, 0]], dtype=float)
```

```python
print("Matrix A:")
print(A)
```

```python
print("\nRank of A:", rank_from_scratch(A))
```

## *Using Libraries*

```python
import numpy as np
```

```python
# ----- Define Matrix -----
A = np.array([[2, 4, 1],
              [6, 12, 3],
              [1, 2, 0]])
```

```python
print("Matrix A:")
print(A)
```

```python
# ----- Find Rank -----
rank = np.linalg.matrix_rank(A)
```

```python
print("\nRank of Matrix A:", rank)
```
