```python
import numpy as np

# ----- Example Data -----
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# ----- Predictions (Assume m & c) -----
m = 0.7
c = 1.5
y_pred = m * X + c

# ----- Mean Squared Error -----
def cost_function(y, y_pred):
    n = len(y)
    cost = (1/(2*n)) * np.sum((y_pred - y)**2)
    return cost

print("Cost (J):", round(cost_function(y, y_pred), 3))
```

```python
import numpy as np

# ----- Example Data -----
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Initialize parameters
m = 0
c = 0
alpha = 0.01  # learning rate
epochs = 1000
n = len(X)

# ----- Gradient Descent -----
for _ in range(epochs):
    y_pred = m * X + c
    dm = (-1/n) * np.sum((y - y_pred) * X)
    dc = (-1/n) * np.sum(y - y_pred)
    m = m - alpha * dm
    c = c - alpha * dc

# ----- Display Results -----
print("Final Slope (m):", round(m, 3))
print("Final Intercept (c):", round(c, 3))
```

```python
import matplotlib.pyplot as plt

m = 0
c = 0
alpha = 0.01
epochs = 1000
n = len(X)
costs = []

for _ in range(epochs):
    y_pred = m * X + c
    cost = (1/(2*n)) * np.sum((y_pred - y)**2)
    costs.append(cost)

    dm = (-1/n) * np.sum((y - y_pred) * X)
    dc = (-1/n) * np.sum(y - y_pred)
    m = m - alpha * dm
    c = c - alpha * dc

plt.plot(range(epochs), costs)
plt.title("Loss Function Over Iterations")
plt.xlabel("Epochs")
plt.ylabel("Cost (MSE)")
plt.show()
```
