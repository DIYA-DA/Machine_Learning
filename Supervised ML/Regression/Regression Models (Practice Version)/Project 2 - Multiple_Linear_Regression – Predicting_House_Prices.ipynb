# 🎯 Goal:
## Predict house prices using multiple features like area, number of bedrooms, and age.

# 📦 Required Libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# 🧾 Sample Dataset:

```python
# Create a simple dataset
data = {
    'Area': [1000, 1500, 1800, 2400, 3000, 3500, 4000, 4100, 4700, 5000],
    'Bedrooms': [2, 3, 3, 4, 4, 4, 5, 5, 5, 6],
    'Age': [10, 15, 20, 5, 8, 12, 7, 3, 4, 2],
    'Price': [200, 250, 280, 350, 400, 450, 500, 520, 580, 600]  # in lakhs
}
df = pd.DataFrame(data)
```

```python
# Display dataset
print(df.head())
```

```python
# Correlation heatmap
sns.heatmap(df.corr(), annot=True, cmap="Blues")
plt.title("Feature Correlation Heatmap")
plt.show()
```

# 🧠 Model Training:

```python
# Features and Target
X = df[['Area', 'Bedrooms', 'Age']]
y = df['Price']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

# 📈 Predict & Evaluate:

```python
# Predict
y_pred = model.predict(X_test)

# Compare predictions
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(comparison)

# Evaluation Metrics
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))

# Regression coefficients
print("\nIntercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

# 📉 Visualization of Predictions:

```python
# Plot Actual vs Predicted
plt.scatter(y_test, y_pred, color='green')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.xlabel("Actual Price (Lakh)")
plt.ylabel("Predicted Price (Lakh)")
plt.title("Actual vs Predicted Prices")
plt.grid(True)
plt.show()
```

# 🧠 Interpretation:
## Intercept: The base price when all features are 0.

## Coefficients: The contribution of each feature to the price.

## A higher R² score indicates a better fit.
